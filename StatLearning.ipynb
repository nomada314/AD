```python
import numpy as np
import numpy as np
import matplotlib
import matplotlib.animation
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
matplotlib.style.use('seaborn')
```

```python
url = "http://faculty.marshall.usc.edu/gareth-james/ISL/Advertising.csv"
ads = pd.read_csv(url)
ads[1:5]
```

```python
fig, axs = plt.subplots(1,3, sharey=True, figsize=(20,10))
axs[0].scatter(ads['TV'], ads['sales'], c="tomato")
axs[1].scatter(ads['radio'], ads['sales'], c="tomato")
axs[2].scatter(ads['newspaper'], ads['sales'], c="tomato")
axs[0].set_xlabel("TV advertising (Thousands of dollars)")
axs[1].set_xlabel("Radio advertising (Thousands of dollars)")
axs[2].set_xlabel("Newspaper advertising (Thousands of dollars)")
axs[0].set_ylabel("Sales (Thousands of units)")


fit    = np.polyfit(ads['TV'], ads['sales'],1)
fit_fn = np.poly1d(fit) 

mn = np.min(ads['TV'])
mx = np.max(ads['TV'])
x1 = np.linspace(mn,mx,500)
axs[0].plot(x1, fit_fn(x1), '-')

fit    = np.polyfit(ads['radio'], ads['sales'],1)
fit_fn = np.poly1d(fit) 

mn = np.min(ads['radio'])
mx = np.max(ads['radio'])
x1 = np.linspace(mn,mx,500)
axs[1].plot(x1, fit_fn(x1), '-')

fit    = np.polyfit(ads['newspaper'], ads['sales'],1)
fit_fn = np.poly1d(fit) 

mn = np.min(ads['newspaper'])
mx = np.max(ads['newspaper'])
x1 = np.linspace(mn,mx,500)
axs[2].plot(x1, fit_fn(x1), '-')

plt.show()
```

```python
mn = np.min(ads['TV'])
mx = np.max(ads['TV'])
x1 = np.linspace(mn,mx,500)

fig = plt.figure()
plt.scatter(ads['TV'], ads['sales'], c="tomato")
plt.xlabel("TV advertising (Thousands of dollars)")
plt.ylabel("Sales (Thousands of units)")

fit    = np.polyfit(ads['TV'], ads['sales'],1)
fit_fn = np.poly1d(fit) 
plt.plot(x1, fit_fn(x1), '-')

fit    = np.polyfit(ads['TV'], ads['sales'],2)
fit_fn = np.poly1d(fit) 
plt.plot(x1, fit_fn(x1), c="darkgreen")


plt.show()
```
